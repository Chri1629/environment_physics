```python
# Libraries

import matplotlib.pyplot as plt
import matplotlib as mpl

import numpy as np

import cartopy.crs as ccrs
import cartopy.feature as cfeature
```

```python
# Example 1
# Explore netCDF files using different methods
```

```python
# Choice of file (model)

modfile='./ts_Amon_IPSL-CM6A-LR_historical_r1i1p1f1_gr_185001-201412.nc'
#modfile='./ts_Amon_BCC-ESM1_historical_r1i1p1f1_gn_185001-201412.nc'

#- extract model name

m_variable='./ts_Amon_'
m_case='_historical_r1i1p1f1_gr_185001-201412.nc'
model=modfile[len(m_variable):-len(m_case)]
```

```python
# Example 1a : ncdump (python script that mimics the command line tool) & netCDF4 library

from ncdump import ncdump
from netCDF4 import Dataset

# Read file with netCDF4 library

d1a = Dataset(modfile, 'r')

# Print attributes with ncdump function

nc_attrs, nc_dims, nc_vars = ncdump(d1a)

# Close

d1a.close()
```

```python
# Example 1b : netCDF4 (alone)

# Explore file with netCDF4 library

d1b = Dataset(modfile, mode='r')
print(d1b.variables.keys() )

# Imoprt data

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
mlons = d1b.variables['lon'][:]
mlats = d1b.variables['lat'][:]
mtime = d1b.variables['time'][:]
mts = d1b.variables['ts'][:,:,:]
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
d1b.close()

nlon=len(mlons)
nlat=len(mlats)
nmonths=len(mtime)

#- check dimensions
print('nlon=',nlon , '; nlat=',nlat , '; nmonths=',nmonths)
print(np.shape(mts)) # > we have a numpy array!
```

```python
# Example 1c : scipy (only works with netCDF3 format)

# Using nco
# Conversion from netCDF4 to netCDF3 can be done in one of the following ways:
# nccopy -k nc3 foo4c.nc foo3.nc
# ncks -3 foo4c.nc foo3.nc

from scipy.io import netcdf

# Import data
d1c = netcdf.netcdf_file('./ts_Amon_IPSL-CM6A-LR_historical_r1i1p1f1_gr_185001-201412.nc3.nc','r')

#Look at the variables available
print(d1c.variables)

#Look at a specific variable's dimensions
print(d1c.variables['ts'].dimensions)   

#Look at a specific variable's units
print(d1c.variables['ts'].units) ## output is ('K')
```

```python
# Example 1d : xarray

import xarray as xr

#- Open file
d1d = xr.open_dataset(modfile)

#- Explore file
print(d1d)

#- extract variable (ts)
ts = d1d.ts  
print('\n','>>>>>>>>>>>>','\n',ts)
```

```python
# Example 2
# Different ways of visualizing raster data (netCDF in this case)

# Let's work with the variable "ts" from the xarray import

# Explore its dimensions > lat/lon appear to contains grid cell limits
print(ts.lat)
print(ts.lon)
```

```python
# Example 3
# Regridding

# Let's assume I want to compare the output of two different model simulations, 
# in terms of the spatial faetures of surface temperature.
# However, these two models use different horizontal grids.
# >> I need spatial interpolation to a common grid >> regridding

# https://climatedataguide.ucar.edu/climate-data-tools-and-analysis/regridding-overview

# https://stackoverflow.com/questions/25544110/regridding-regular-netcdf-data
# http://xarray.pydata.org/en/stable/interpolation.html#example

# I will proceed using xarray
```

```python
# My first model is the one used above (IPSL-CM6A-LR)
# We had: nlon= 144 ; nlat= 143 ; nmonths= 1980

# Let's get the tempartaure from a second model

#- extract model name
modfile2='./ts_Amon_MRI-ESM2-0_historical_r1i1p1f1_gn_185001-201412.nc'
#modfile2='./ts_Amon_BCC-ESM1_historical_r1i1p1f1_gn_185001-201412.nc'

m_variable='./ts_Amon_'
m_case='_historical_r1i1p1f1_gr_185001-201412.nc'
model2=modfile2[len(m_variable):-len(m_case)]

#- Open file
m2 = xr.open_dataset(modfile2)

#- extract variable (ts)
ts2 = m2.ts  
print(ts2)

ts2.mean(axis=0).plot()
```

```python
# Apply regridding

ts2i = ts2.interp(lat=ts.lat, lon=ts.lon, method='linear')

# Compare to original

print(ts2i)
ts2i.mean(axis=0).plot()
```

```python
# Apply regridding

ts2in = ts2.interp(lat=ts.lat, lon=ts.lon, method='nearest')

# Compare to original

print(ts2in)
ts2in.mean(axis=0).plot()
```

```python
# Compare the bilinear and nearest neighbour methods

intdiff = (ts2in.mean(axis=0)-ts2i.mean(axis=0))/ts2i.mean(axis=0)   # relative difference
intdiff.plot()
```

```python
# Now compare the two models

# Switch from Kelvin to Celsius degrees

ts_avg=ts.mean(axis=0)-273.15
ts2_avg=ts2.mean(axis=0)-273.15
ts2i_avg=ts2i.mean(axis=0)-273.15

# Set plot

fig = plt.figure(figsize=(13,8))  # x,y(inches)

subplots = (2,2)
n_panels = subplots[0] * subplots[1]

# I will use the same colormap and scale for all plots, to facilitate comparison.
# Because we will also show the difference, we want this scale to be symmetric about zero

tmax = 40
norm = mpl.colors.Normalize(vmin=-tmax,vmax=tmax) 
cmap = mpl.cm.seismic

# Model 1

ax = fig.add_subplot(subplots[0], subplots[1], 1, projection=ccrs.PlateCarree())
ax.set_title(''.join([model,'\n']))
ax.set_global()
mm = ax.pcolormesh(lon_cyc, ts.lat, ts_avg, \
                   transform=ccrs.PlateCarree(),cmap=cmap, norm=norm )  
ax.coastlines()
ax.gridlines(draw_labels=True)


# Model 2 (orig)

ax = fig.add_subplot(subplots[0], subplots[1], 2, projection=ccrs.PlateCarree())
ax.set_title(''.join([model2,' (original)\n']))
ax.set_global()
mm = ax.pcolormesh(ts2.lon, ts2.lat, ts2_avg, \
                   transform=ccrs.PlateCarree(),cmap=cmap, norm=norm )  
ax.coastlines()
ax.gridlines(draw_labels=True)

# Model 2 (regr)

ax = fig.add_subplot(subplots[0], subplots[1], 4, projection=ccrs.PlateCarree())
ax.set_title(''.join([model2,' (regridded)\n']))
ax.set_global()
mm = ax.pcolormesh(lon_cyc, ts.lat, ts2i_avg, \
                   transform=ccrs.PlateCarree(),cmap=cmap, norm=norm )  
ax.coastlines()
ax.gridlines(draw_labels=True)


# Model 1 - Model 2 (regr)

tsdiff_avg = (ts2i_avg-ts_avg)  # absolute difference
tsdiff_avg_cyc, lon_cyc = add_cyclic_point(tsdiff_avg, coord=ts.lon) 

ax = fig.add_subplot(subplots[0], subplots[1], 3, projection=ccrs.PlateCarree())

ax.set_title(''.join([model,' - ',model2,'\n']))
ax.set_global()
mm = ax.pcolormesh(lon_cyc, ts.lat, tsdiff_avg_cyc, \
                   transform=ccrs.PlateCarree(),cmap=cmap, norm=norm )  
ax.coastlines()
ax.gridlines(draw_labels=True)


#- add colorbar
cbar_ax = fig.add_axes([0.95, 0.30, 0.02, 0.40]) #[left, bottom, width, height]
cbar = fig.colorbar(mm, cax=cbar_ax, extend='both', orientation='vertical')
cbar.set_label('Celsius')
cbar.ax.tick_params(labelsize=8)

plt.show()
plt.close()
```

```python
# Apply regridding

ts3i = ts3.interp(lat=ts.lat, lon=ts.lon, method='linear')
ts3i.mean(axis=0).plot()
ts3i_avg = ts3i.mean(axis=0)-273.15
```

```python
# Module 3 (assignment)
# Compare the time series of annual GMTA (global mean temperature anomaly) from a multi-model ensemble with the corresponding observational time series (HadCRUT4) since 1901.
# You can build on the output of examples from modules 2 and 3.
```

```python
# First, let's calculate the ensemble mean on the whole time series for the last "ny" years

ny = 114

# First, let's do a time slicing & aggregate in time to annual resolution

tsy = np.empty([ny,len(ts.lat),len(ts.lon),3],dtype=float)
print(np.shape(tsy))

y0=0
for yi in range(0,ny):
  tsy[yi,:,:,0] = np.nanmean(ts[-(ny*12)+y0:-(ny*12)+y0+12,:,:],axis=0) 
  tsy[yi,:,:,1] = np.nanmean(ts2i[-(ny*12)+y0:-(ny*12)+y0+12,:,:],axis=0)
  tsy[yi,:,:,2] = np.nanmean(ts3i[-(ny*12)+y0:-(ny*12)+y0+12,:,:],axis=0)    
  y0+=12

# Second, let's go from absolute values to anomalies 
# Here I am making two approximations, which makes it different from the observational time series we already know
# 1. I am calculating the anomaly with respect to the entire time series, and not 1961-1990
# 2. I am calculating the anomaly on the annual (rather than monthly) time series

tsy = tsy - np.nanmean(tsy,axis=0)
print(np.shape(np.nanmean(tsy,axis=0)))
print(np.shape(tsy))
```

```python
# Third, aggregate the 3 models: multi-model average

import datetime
print(datetime.datetime.now().time())

ens_mean_ts = np.nanmean(tsy,axis=3) #-273.15

# This other way takes forever ... 7 minutes just for 1 year
#ens_mean_ts = np.nanmean(np.array([ts[-(ny*12):,:,:],ts2i[-(ny*12):,:,:],ts3i[-(ny*12):,:,:]]),axis=0)

print(datetime.datetime.now().time())
print(np.shape(ens_mean_ts))

print(np.nanmean(ens_mean_ts))
```

```python
# Set plot 
# --> only makes sense if the curve represents the absolute value, not the anomaly, whose average is 0 by definition

fig = plt.figure(figsize=(12,12))  # x,y(inches)

subplots = (1,1)
n_panels = subplots[0] * subplots[1]

# I will use the same colormap and scale for all plots, to facilitate comparison.
# Because we will also show the difference, we want this scale to be symmetric about zero

tmax = 40
norm = mpl.colors.Normalize(vmin=-tmax,vmax=tmax) 
cmap = mpl.cm.seismic

# Mean

ax = fig.add_subplot(subplots[0], subplots[1], 1, projection=ccrs.PlateCarree())
ax.set_title(' '.join(['Ensemble mean:',model,model2,model3,'\n']))
ax.set_global()
mm = ax.pcolormesh(lon_cyc, ts.lat, np.nanmean(ens_mean_ts,axis=0), \
                   transform=ccrs.PlateCarree(),cmap=cmap, norm=norm )  
ax.coastlines()
ax.gridlines(draw_labels=True)

#- add colorbar
cbar_ax = fig.add_axes([0.95, 0.35, 0.02, 0.30]) #[left, bottom, width, height]
cbar = fig.colorbar(mm, cax=cbar_ax, extend='both', orientation='vertical')
cbar.set_label('Celsius')
cbar.ax.tick_params(labelsize=8)

plt.show()
plt.close()
```

```python
# Fourth: now we need spatial interpolation

# 1. Let's exploit what we did in Module 1 and calculate the areas of grid cells here

area = np.empty([len(ts.lat),len(ts.lon)],dtype=float)

er = 6371 # km
from math import sin,radians,pi

for j in range(0,len(ts.lat)-1):
  area[j,:]= 2*pi*(er**2) * abs(sin(radians(ts.lat[j]))-sin(radians(ts.lat[j+1]))) * len(ts.lon)/360
  #print(np.array(ts.lat[j]),np.array(ts.lat[j+1]),np.array(ts[0,j,0]),np.array(ts[0,j+1,0]),area[j,0])
area[-1,:]=area[0,:]
    
fig = plt.figure(figsize=(9,6))  # x,y(inches)

ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_title(' '.join(['Grid cell area','\n']))
ax.set_global()
mm = ax.pcolormesh(lon_cyc, ts.lat, area/1e5, \
                   transform=ccrs.PlateCarree(),cmap=cmap, norm=norm )  
ax.coastlines()
ax.gridlines(draw_labels=True)

#- add colorbar
cbar_ax = fig.add_axes([0.28, 0.10, 0.46, 0.05]) #[left, bottom, width, height]
cbar = fig.colorbar(mm, cax=cbar_ax, extend='both', orientation='horizontal')
cbar.set_label('$1e5$ $m^{2}$')
cbar.ax.tick_params(labelsize=8)

plt.show()
plt.close()
```

```python
# 2. Let's make an area-weighted global average temperature anomaly time series (in two steps)

ens_time_series_lats = np.average(ens_mean_ts, weights=area[:,0], axis=1)
print(np.shape(ens_time_series_lats))

ens_time_series = np.nanmean(ens_time_series_lats,axis=1) 
print(np.shape(ens_time_series))
```

```python
# Let's read the observational data and extract 1900-2014

import pandas as pd
import datetime

gmta = pd.read_csv('./HadCRUT4.txt',sep='\t',header=None,names=['time','obs'])
gmta['time'] = pd.to_datetime(gmta['time'], format="%Y")
gmta = gmta.set_index(['time'])

gmta.drop(gmta.loc[gmta.index <= pd.to_datetime('1900-01-01')].index,inplace=True)
gmta.drop(gmta.loc[gmta.index > pd.to_datetime('2014-01-01')].index,inplace=True)
print(gmta)
```

```python
# Add multi-model ensemble to gmta & plot

gmta['model'] = ens_time_series

gmta.plot()
```
