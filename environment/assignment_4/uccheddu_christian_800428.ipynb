We choose a model that contains both temperatures and pressures so that we can do the work on both quantities. We will then draw everything in parallel so that we can see both at the same time. In this case we have chose **MRI-ESM2-0** model.

```python
model = 'MRI-ESM2-0'
```

We read the data and extrapolate the historical series regarding pressure and temperature.

```python
modfile0=''.join(['dati/psl_Amon_',model,'_historical_r1i1p1f1_gn_185001-201412.nc'])
modfile1=''.join(['dati/ts_Amon_',model,'_historical_r1i1p1f1_gn_185001-201412.nc'])

ds0 = xr.open_dataset(modfile0)
ds1 = xr.open_dataset(modfile1)

psl = ds0.psl
ts = ds1.ts
```

Let's first show the average value of the time series on the map to see if the models have been imported correctly.

```python
psl_avg_cyc, psl_lon_cyc = add_cyclic_point(psl.mean(axis=0), coord=psl.lon)  
ts_avg_cyc, ts_lon_cyc = add_cyclic_point(ts.mean(axis=0) - 273.15, coord=ts.lon)  


fig = plt.figure(figsize=(13,8))
gs = gridspec.GridSpec(nrows=1, ncols=2, height_ratios=[1], hspace = 0.2)
cmap1 = mpl.cm.jet
cmap2 = mpl.cm.coolwarm

ax0 = fig.add_subplot(gs[0], projection=ccrs.PlateCarree())
ax0.set_title('Pressure {}'.format(model))
ax0.set_global()
mm0 = ax0.pcolormesh(psl_lon_cyc, psl.lat, psl_avg_cyc, \
                   transform=ccrs.PlateCarree(),cmap=cmap1)  

ax0.coastlines()
ax0.gridlines(draw_labels=True)
cbar_ax0 = fig.add_axes([0.05, 0.35, 0.02, 0.30]) #[left, bottom, width, height]
cbar0 = fig.colorbar(mm0, cax=cbar_ax0, extend='both', orientation='vertical')
cbar0.set_label('Pressure (Pa)')
cbar0.ax.tick_params(labelsize=8)
cbar_ax0.yaxis.tick_left()


norm2 = mpl.colors.Normalize(vmin=-40,vmax=40) 

ax1 = fig.add_subplot(gs[1], projection=ccrs.PlateCarree())
ax1.set_title('Temperature {}'.format(model))
ax1.set_global()
mm1 = ax1.pcolormesh(ts_lon_cyc, ts.lat, ts_avg_cyc, \
                   transform=ccrs.PlateCarree(),cmap=cmap2, norm = norm2 )  

ax1.coastlines()
ax1.gridlines(draw_labels=True)

#- add colorbar
cbar_ax1 = fig.add_axes([0.95, 0.35, 0.02, 0.30]) #[left, bottom, width, height]
cbar1 = fig.colorbar(mm1, cax=cbar_ax1, extend='both', orientation='vertical')
cbar1.set_label('Temperature (Celsius)')
cbar1.ax.tick_params(labelsize=8)
```

We then calculate the anomalies with respect to the required periods. In particular, therefore, we extrapolate both for the historical series of temperatures and for that of pressures the two sub-series corresponding to the period 1900-1950 and 1950-2000. **Then we calculate the time average of these time series and then we make the difference in order to obtain the anomalies between the two periods.**

```python
psl_first = psl.loc["1900":"1950"].mean(dim = "time")
psl_last = psl.loc["1950":"2000"].mean(dim = "time")

ts_first = ts.loc["1900":"1950"].mean(dim = "time")
ts_last = ts.loc["1950":"2000"].mean(dim = "time")
```

```python
ts_anomaly = ts_last - ts_first
psl_anomaly = psl_last - psl_first
```

We draw on the map by placing the anomalies of pressures and temperatures side by side.

```python
psl_avg_cyc, psl_lon_cyc = add_cyclic_point(psl_anomaly, coord=psl.lon)  
ts_avg_cyc, ts_lon_cyc = add_cyclic_point(ts_anomaly, coord=ts.lon)  

norm1 = mpl.colors.Normalize(vmin=-135,vmax=135) 
fig = plt.figure(figsize=(13,8))
gs = gridspec.GridSpec(nrows=1, ncols=2, height_ratios=[1], hspace = 0.2)
cmap1 = mpl.cm.jet
cmap2 = mpl.cm.coolwarm

ax0 = fig.add_subplot(gs[0], projection=ccrs.PlateCarree())
ax0.set_title('Pressure {}'.format(model))
ax0.set_global()
mm0 = ax0.pcolormesh(psl_lon_cyc, psl.lat, psl_avg_cyc, \
                   transform=ccrs.PlateCarree(),cmap=cmap1,  norm = norm1)  

ax0.coastlines()
ax0.gridlines(draw_labels=True)
cbar_ax0 = fig.add_axes([0.05, 0.35, 0.02, 0.30]) #[left, bottom, width, height]
cbar0 = fig.colorbar(mm0, cax=cbar_ax0, extend='both', orientation='vertical')
cbar0.set_label('Pressure (Pa)')
cbar0.ax.tick_params(labelsize=8)
cbar_ax0.yaxis.tick_left()


norm2 = mpl.colors.Normalize(vmin=-1.5,vmax=1.5) 

ax1 = fig.add_subplot(gs[1], projection=ccrs.PlateCarree())
ax1.set_title('Temperature {}'.format(model))
ax1.set_global()
mm1 = ax1.pcolormesh(ts_lon_cyc, ts.lat, ts_avg_cyc, \
                   transform=ccrs.PlateCarree(),cmap=cmap2, norm = norm2)  

ax1.coastlines()
ax1.gridlines(draw_labels=True)

#- add colorbar
cbar_ax1 = fig.add_axes([0.95, 0.35, 0.02, 0.30]) #[left, bottom, width, height]
cbar1 = fig.colorbar(mm1, cax=cbar_ax1, extend='both', orientation='vertical')
cbar1.set_label('Temperature (Celsius)')
cbar1.ax.tick_params(labelsize=8)
```
