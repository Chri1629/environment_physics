Import of the libraries

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

import cartopy.crs as ccrs
from cartopy.util import add_cyclic_point

import xarray as xr
```

Selection of the model

```python
model = 'MRI-ESM2-0'
```

Reading and importing models, one for pressure and one for temperature

```python
modfile0=''.join(['./psl_Amon_',model,'_historical_r1i1p1f1_gn_185001-201412.nc'])
modfile1=''.join(['./ts_Amon_',model,'_historical_r1i1p1f1_gn_185001-201412.nc'])

ds0 = xr.open_dataset(modfile0)
ds1 = xr.open_dataset(modfile1)

psl = ds0.psl
ts = ds1.ts
```

## Pressione

First plotting the average pressure accross all years

```python
plt.rcParams['figure.figsize'] = (9,6)
```

```python
psl_avg_cyc, lon_cyc = add_cyclic_point(psl.mean(axis=0), coord=psl.lon)  

fig = plt.figure()

ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_title(''.join([model,'\n']))
ax.set_global()
mm = ax.pcolormesh(lon_cyc, psl.lat, psl_avg_cyc, \
                   transform=ccrs.PlateCarree(),cmap=mpl.cm.bwr )  

ax.coastlines()
ax.gridlines(draw_labels=True)

#- add colorbar
cbar_ax = fig.add_axes([0.28, 0.10, 0.46, 0.05]) #[left, bottom, width, height]
cbar = fig.colorbar(mm, cax=cbar_ax, extend='both', orientation='horizontal')
cbar.set_label('Pressure')
cbar.ax.tick_params(labelsize=8)

plt.show()
plt.close()
```

Now computing the anomalies with respect to the global mean of all years

```python
psl_anomalies = psl - psl.mean(dim='time')
```

Now saving the periods of interest

```python
psl_anom_first = psl_anomalies.loc["1900":"1950"]
psl_anom_last = psl_anomalies.loc["1950":"2000"]
```

```python
# come si fa il subplots????????

fig, (ax1,ax2) = plt.subplots(1, 2)

psl_avg_cyc1, lon_cyc1 = add_cyclic_point(psl_anom_first.mean(axis=0), coord=psl.lon)  

ax1 = plt.axes(projection=ccrs.PlateCarree())
ax1.set_title(''.join([model,'\n']))
ax1.set_global()
mm1 = ax1.pcolormesh(lon_cyc1, psl.lat, psl_avg_cyc1, \
                   transform=ccrs.PlateCarree(),cmap=mpl.cm.bwr )  

ax1.coastlines()
ax1.gridlines(draw_labels=True)

#- add colorbar
cbar_ax = fig.add_axes([0.28, 0.10, 0.46, 0.05]) #[left, bottom, width, height]
cbar = fig.colorbar(mm1, cax=cbar_ax, extend='both', orientation='horizontal')
cbar.set_label('Pressure anomalies')
cbar.ax.tick_params(labelsize=8)

psl_avg_cyc2, lon_cyc2 = add_cyclic_point(psl_anom_last.mean(axis=0), coord=psl.lon)  

ax2 = plt.axes(projection=ccrs.PlateCarree())
ax2.set_title(''.join([model,'\n']))
ax2.set_global()
mm2 = ax2.pcolormesh(lon_cyc2, psl.lat, psl_avg_cyc2, \
                   transform=ccrs.PlateCarree(),cmap=mpl.cm.bwr )  

ax2.coastlines()
ax2.gridlines(draw_labels=True)

#- add colorbar
cbar_ax = fig.add_axes([0.28, 0.10, 0.46, 0.05]) #[left, bottom, width, height]
cbar = fig.colorbar(mm2, cax=cbar_ax, extend='both', orientation='horizontal')
cbar.set_label('Pressure anomalies')
cbar.ax.tick_params(labelsize=8)
```

```python
first = psl.loc["1900":"1950"]
last = psl.loc["1950":"2000"]
```

```python
diff = abs(first.mean(axis=0) - last.mean(axis=0))
```

```python
from cartopy.util import add_cyclic_point
psl_avg_cyc, lon_cyc = add_cyclic_point(diff, coord=psl.lon)  

fig = plt.figure(figsize=(9,6))

ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_title(''.join([model,'\n']))
ax.set_global()
mm = ax.pcolormesh(lon_cyc, psl.lat, psl_avg_cyc, \
                   transform=ccrs.PlateCarree(),cmap=mpl.cm.jet )  

ax.coastlines()
ax.gridlines(draw_labels=True)

#- add colorbar
cbar_ax = fig.add_axes([0.28, 0.10, 0.46, 0.05]) #[left, bottom, width, height]
cbar = fig.colorbar(mm, cax=cbar_ax, extend='both', orientation='horizontal')
cbar.set_label('Absolute pressure differences')
cbar.ax.tick_params(labelsize=8)

plt.show()
plt.close()
```

## Temperature

```python
ts_avg_cyc, lon_cyc = add_cyclic_point(ts.mean(axis=0) - 273.15, coord=ts.lon)  

tmax = 50
norm = mpl.colors.Normalize(vmin=-tmax,vmax=tmax) 

fig = plt.figure(figsize=(9,6))

ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_title(''.join([model,'\n']))
ax.set_global()
mm = ax.pcolormesh(lon_cyc, ts.lat, ts_avg_cyc, \
                   transform=ccrs.PlateCarree(),cmap=mpl.cm.coolwarm, norm = norm )  

ax.coastlines()
ax.gridlines(draw_labels=True)

#- add colorbar
cbar_ax = fig.add_axes([0.28, 0.10, 0.46, 0.05]) #[left, bottom, width, height]
cbar = fig.colorbar(mm, cax=cbar_ax, extend='both', orientation='horizontal')
cbar.set_label('Temperature')
cbar.ax.tick_params(labelsize=8)

plt.show()
plt.close()
```
