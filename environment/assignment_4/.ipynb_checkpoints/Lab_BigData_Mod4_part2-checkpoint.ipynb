```python
# Libraries

import matplotlib.pyplot as plt
import matplotlib as mpl

import numpy as np

import cartopy.crs as ccrs

import xarray as xr
```

```python
# Example 1
# Calculate and display the EOF-NAO index from one of the CMIP6 historical simulations
```

```python
# Chose models

model = 'MRI-ESM2-0'
```

```python
#-------------------------------------------------
#-- Read and organize Sea Level Pressure data ----
#-------------------------------------------------

print('\n **** NOW ARRANGE SLP *** \n')

#- Path to file
modfile0=''.join(['./psl_Amon_',model,'_historical_r1i1p1f1_gn_185001-201412.nc'])

#- explore file
ds0 = xr.open_dataset(modfile0)

#- extract variable (psl) - temporary variable
psl0 = ds0.psl
print(ds0)
```

```python
#- make some visual checks of what's going on
fig = plt.figure(figsize=(16,12))
subplots = (2,2)
n_panels = subplots[0] * subplots[1]

#- check: location before lon adjustment
ax = fig.add_subplot(subplots[0], subplots[1], 1, projection=ccrs.PlateCarree())
ax.set_global()
ax.coastlines() 
psl0m = psl0.mean(axis=0)
psl0m.plot(extend='both')
```

```python
# ~~~ Remember ~~~
# SLP anomalies over the Atlantic sector, 20°-80°N, 90°W-40°E. 

#- focus on the region of interest 
ds0_nat = ds0.sel(lat=slice(20,80), lon=slice(-90,40))
psl_nat=ds0_nat.psl

#- check
fig = plt.figure(figsize=(9,6))  # x,y(inches)

ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_title(''.join([model,'\n']))
ax.set_global()
mm = ax.contourf(psl_nat.lon, psl_nat.lat, psl_nat.mean(axis=0), \
                   transform=ccrs.PlateCarree(),cmap=mpl.cm.jet )   

ax.coastlines()
ax.gridlines(draw_labels=True)

#- add colorbar
cbar_ax = fig.add_axes([0.28, 0.10, 0.46, 0.05]) #[left, bottom, width, height]
cbar = fig.colorbar(mm, cax=cbar_ax, extend='both', orientation='horizontal')
cbar.set_label('Pa')
cbar.ax.tick_params(labelsize=8)

plt.show()
plt.close()
```

```python
#- Adjust longitudes

#- (Option 1: in python)
ds0.coords['lon'] = (ds0.coords['lon'] + 180) % 360 - 180
ds0 = ds0.sortby(ds0.lon)

#- (Option 2: with nco from shell)
# ncap2 -O -s 'where(lon>180) lon=lon-360' ifile ofile
```

```python
# re-try after fixing longitudes

#- focus on the region of interest 
ds0_nat = ds0.sel(lat=slice(20,80), lon=slice(-90,40))
psl_nat=ds0_nat.psl

#- check
fig = plt.figure(figsize=(9,6))  # x,y(inches)

ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_title(''.join([model,'\n']))
ax.set_global()
mm = ax.contourf(psl_nat.lon, psl_nat.lat, psl_nat.mean(axis=0), \
                   transform=ccrs.PlateCarree(),cmap=mpl.cm.jet )   

ax.coastlines()
ax.gridlines(draw_labels=True)

#- add colorbar
cbar_ax = fig.add_axes([0.28, 0.10, 0.46, 0.05]) #[left, bottom, width, height]
cbar = fig.colorbar(mm, cax=cbar_ax, extend='both', orientation='horizontal')
cbar.set_label('Pa')
cbar.ax.tick_params(labelsize=8)

plt.show()
plt.close()
```

```python
#- focus on the season of interest (boreal winter)

def djfm(month):
    return ((month == 1) | (month == 2) | (month == 3) | (month == 12))

psl_nat_m = psl_nat.sel(time=djfm(psl_nat['time.month']))
psl_nat_ms = psl_nat_m.coarsen(time=4).mean()
```

```python
#-- Calculate and plot NAO index based on EOFs ---

# Compute anomalies by removing the time-mean.
psl_anom = psl_nat_ms - psl_nat_ms.mean(dim='time')

from eofs.xarray import Eof

# Retrieve the leading EOF, expressed as the covariance (alternatives: e.g. correlation) 
# between the leading PC time series and the input SST anomalies at each grid point, and the
# leading PC time series itself.

solver = Eof(psl_anom)
eof1 = solver.eofsAsCovariance(neofs=1)
pc1 = solver.pcs(npcs=1, pcscaling=1)
```

```python
#- set multi-panel plot
fig = plt.figure(figsize=(8,12))
subplots = (2,1)
n_panels = subplots[0] * subplots[1]

#- plot leading EOF: spatial component of the NAO index
ax = fig.add_subplot(subplots[0], subplots[1], 1, projection=ccrs.PlateCarree())
ax.set_global()
ax.coastlines() 
eof1[0,:,:].plot()

#- plot leading EOF time series: temporal component of the NAO index
ax = fig.add_subplot(subplots[0], subplots[1], 2)
pc1.plot(marker='.')

#- Store/visualize plot
plt.show()
plt.close()
```

```python
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
# Task 2. 
# Calculate the Southern Oscillation Index at monthly resolution 
# for two models of your choice for the period 1950-1999; 
# plot time series
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
```

```python
#- Chose models
model=['MRI-ESM2-0','MIROC6']
cols=['blue','red']
```

```python
#- Define coordinates of Darwin and Tahiti

lon_Tah = -149.45
lat_Tah = -17.67

lon_Dar = 130.84
lat_Dar = -12.44
```

```python
#- Set plot
fig = plt.figure(figsize=(8,12))
subplots = (4,1)
n_panels = subplots[0] * subplots[1]

#- plot station locations
ax = fig.add_subplot(subplots[0], subplots[1], 1, projection=ccrs.PlateCarree())
ax.set_global()
ax.coastlines() 
plt.plot([lon_Tah], [lat_Tah],
         color='pink', linewidth=2, marker='*', markersize=10, markeredgecolor='black',
         transform=ccrs.Geodetic())
plt.plot([lon_Dar], [lat_Dar],
         color='cyan', linewidth=2, marker='*', markersize=10, markeredgecolor='black',
         transform=ccrs.Geodetic())

#import nc_time_axis
#import datetime as dtm
#from datetime import datetime

#- Now loop over the two models
for i in (0,1):

  #- Path to file
  modfile1=''.join(['./psl_Amon_',model[i],'_historical_r1i1p1f1_gn_185001-201412.nc'])

  #- explore file
  ds2 = xr.open_dataset(modfile1)
  print(ds2)
  
  #- check longitude range and if needed adjust to (-180,180) E
  if (ds2.lon.max() > 180):
    print('Adjusting longitudes')
    ds2.coords['lon'] = (ds2.coords['lon'] + 180) % 360 - 180
    ds2 = ds2.sortby(ds2.lon)
  
  #- extract variable (psl)
  psl2 = ds2.psl / 100.  # Pa to hPa
  
  #- focus on the locations of interest 
  psl_Tah = psl2.sel(lon=lon_Tah, lat=lat_Tah, method='nearest')
  psl_Dar = psl2.sel(lon=lon_Dar, lat=lat_Dar, method='nearest')
  
  #- focus on the period of interest 
  psl_Tah_y = psl_Tah.sel(time=slice("1950-01", "1999-12"))  
  psl_Dar_y = psl_Dar.sel(time=slice("1950-01", "1999-12")) 


  #- plot time series for the current model
  ax = fig.add_subplot(subplots[0], subplots[1], i+2) 
  psl_Tah_y.plot(color='pink')
  psl_Dar_y.plot(color='cyan')
 
  #- Calculate standardized Tahiti
  st_Tah = ( psl_Tah_y - psl_Tah_y.mean() ) / psl_Tah_y.std()
 
  #- Calculate standardized Darwin
  st_Dar = ( psl_Dar_y - psl_Dar_y.mean() ) / psl_Dar_y.std()

  #- Calculate SOI
  MSD = (st_Tah-st_Dar).std()
  SOI = st_Tah-st_Dar / MSD

  #- plot SOI for current model
  ax = fig.add_subplot(subplots[0], subplots[1], 4) 
  SOI.plot(color=cols[i])

plt.subplots_adjust(hspace=0.6) 
plt.show()  
plt.close()
```
