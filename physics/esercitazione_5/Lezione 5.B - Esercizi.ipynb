## Cluster e elbow method

Proviamo a predere un dataset con tutti i paesi e la loro latitudine e longitudine e a effettuarne una clusterizzazione:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = pd.read_csv('Countries-exercise.csv')
data.head()
```

Proviamo prima a disegnare uno scatter plot per vedere come si distribuiscono

```python
fig = plt.figure(figsize = (8,8))
sns.scatterplot(data=data, x="Longitude", y="Latitude", palette = "plasma_r")
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.grid()
```

Facciamo del clustering non supervisionato applicando il metodo Silhouette e il metodo elbow per vedere qual è il numero migliore di cluster all'interno del dataset

```python
from sklearn.cluster import KMeans
x = data[['Latitude', 'Longitude']]
kmeans = KMeans(5)
# Fit the input data, i.e. cluster the data in X in K clusters
kmeans.fit(x)
```

```python
identified_clusters = kmeans.predict(x)
# Check the result
identified_clusters
```

```python
data_with_clusters = data.copy()
# Create a new Series, containing the identified cluster for each observation
data_with_clusters['Cluster'] = identified_clusters
```

```python
data_with_clusters.loc[data_with_clusters['Cluster'] == 4]
```

```python
sns.scatterplot(data=data_with_clusters, x="Longitude", y="Latitude", hue = "Cluster")
```

```python
wcss=[]

# Create all possible cluster solutions with a loop
for i in range(1,7):
    # Cluster solution with i clusters
    kmeans = KMeans(i)
    # Fit the data
    kmeans.fit(x)
    # Find WCSS for the current iteration
    wcss_iter = kmeans.inertia_
    # Append the value to the WCSS list
    wcss.append(wcss_iter)
    
wcss
```

```python
number_clusters = range(1,7)
plt.plot(number_clusters,wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Within-cluster Sum of Squares')
```

Sembra che il numero giusto di cluster sia 3.

```python
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_samples, silhouette_score

range_clusters = enumerate([2, 3, 4, 5, 6])
for i, k in range_clusters :
        fig, (ax1, ax2) = plt.subplots(1, 2)
        fig.set_size_inches(18, 7)

        km = KMeans(n_clusters=k)
        km_fit = km.fit(x)
        labels = km.predict(x)

        centroids = km.cluster_centers_
        
        # Get silhouette samples
        silhouette_vals = silhouette_samples(x, labels)

        # Silhouette plot
        y_ticks = []
        y_lower, y_upper = 0, 0

        for i, cluster in enumerate(np.unique(labels)):
            cluster_silhouette_vals = silhouette_vals[labels == cluster]
            cluster_silhouette_vals.sort()

            y_upper += len(cluster_silhouette_vals)
            ax1.barh(range(y_lower, y_upper), cluster_silhouette_vals, edgecolor='none', height=1)
            ax1.text(-0.03, (y_lower + y_upper) / 2, str(i + 1))
            y_lower += len(cluster_silhouette_vals)
            
        print("For n_clusters =", k, "The average silhouette_score is :", np.mean(silhouette_vals)) 

        # Get the average silhouette score and plot it
        avg_score = np.mean(silhouette_vals)
        ax1.axvline(avg_score, linestyle='--', linewidth=2, color='green')
        ax1.set_yticks([])
        ax1.set_xlim([-0.1, 1])
        ax1.set_xlabel('Silhouette coefficient values')
        ax1.set_ylabel('Cluster labels')
        ax1.set_title('Silhouette plot for the various clusters', y=1.02);


        ax2.scatter(x['Longitude'], x['Latitude'], marker='.', c=labels, cmap='rainbow', alpha=0.3) #print data
        ax2.scatter(centroids[:, 0], centroids[:, 1], marker='*', c='r', s=250) #print centroids
        ax2.set_xlabel('height')
        ax2.set_ylabel('weight')
        ax2.set_title('Visualization of clustered data', y=1.02)
        plt.suptitle(f'Silhouette analysis using k = {k}', fontsize=16, fontweight='semibold', y=1.05);
```
