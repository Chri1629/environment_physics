# Violin

In questo disegniamo i violin finalmente puliti

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np 
import seaborn as sns
import math
from scipy.stats import pearsonr
import scipy.stats as stats

print("Pandas version: ", pd.__version__)
print("Matplotlib version: ", matplotlib.__version__)
print("Seaborn version: ", sns.__version__)
print("Numpy version: ", np.__version__)

pd.options.mode.chained_assignment = None
```

```python
dati = pd.read_csv("dati_corretti/1_rilevazione_nulli_riempiti_no_outlier.csv")
cvi = pd.read_csv("dati_corretti/cvi_finali.csv")
people = pd.read_csv("dati_corretti/people.csv", sep = ";")
```

```python
dati = pd.merge(people[['id', "Age", "BMI","gender","menopausal"]], dati, on = "id")
cvi = pd.merge(people[['id', "Age", "BMI","gender","menopausal"]], cvi, on = "id")
```

```python
uomini = pd.merge(people.loc[people['menopausal'] == 'male'][['id', "Age", "BMI","gender","menopausal"]], dati, on = "id")
donne = pd.merge(people.loc[people['menopausal'] == 'female'][['id', "Age", "BMI","gender","menopausal"]], dati, on = "id")
donne_and_meno = pd.merge(people.loc[people['gender'] == 'f'][['id', "Age", "BMI", "gender","menopausal"]], dati, on = "id")
menopausal = pd.merge(people.loc[people['menopausal'] == 'menopausal female'][['id', "Age", "BMI","gender","menopausal"]], dati, on = "id")
```

```python
uomini_cvi = pd.merge(people.loc[people['menopausal'] == 'male'][['id', "Age", "BMI","gender","menopausal"]], cvi, on = "id")
donne_cvi = pd.merge(people.loc[people['menopausal'] == 'female'][['id', "Age", "BMI","gender","menopausal"]], cvi, on = "id")
donne_and_meno_cvi = pd.merge(people.loc[people['gender'] == 'f'][['id', "Age", "BMI", "gender","menopausal"]], cvi, on = "id")
menopausal_cvi = pd.merge(people.loc[people['menopausal'] == 'menopausal female'][['id', "Age", "BMI","gender","menopausal"]], cvi, on = "id")
```

## TWO SPLITTED MEDI

Facciamo quelli splittati solo per uomini e donne sui valori medi

```python
colonna = dati.columns[17]

fig= plt.figure(figsize = (10,6))

sns.kdeplot(data=dati, x=dati[colonna],  hue=dati['gender'], shade = True, palette = ["blue", "pink"])

plt.vlines(x = uomini[colonna].median(), ymin = 0, ymax = 0.032, label = "Median men", linestyles = "--", color = "black")
plt.vlines(x = donne_and_meno[colonna].median(), ymin = 0, ymax = 0.065, label = "Median women", linestyles = "--", color = "black")

plt.xticks(size = 12)
plt.yticks(size = 12)
#plt.legend(frameon=False)
plt.legend([],[], frameon=False)
plt.xlabel("Values", fontsize = 15)
plt.ylabel("Density of probability", fontsize = 15)
plt.title("{}".format(colonna), fontsize = 20)
fig.savefig("pics_finali/violin/two_split/medi/{}.png".format(colonna), dpi = 300)
```

```python
colonna = dati.columns[16]

fig= plt.figure(figsize = (10,6))

sns.kdeplot(data=dati, x=dati[colonna], hue=dati['menopausal'], shade = True, palette = ["blue", "violet", "pink"])

plt.vlines(x = uomini[colonna].median(), ymin = 0, ymax = 0.02, label = "Median men", linestyles = "--", color = "black")
plt.vlines(x = donne[colonna].median(), ymin = 0, ymax = 0.030, label = "Median women", linestyles = "--", color = "black")
plt.vlines(x = menopausal[colonna].median(), ymin = 0, ymax = 0.0163, label = "Median menopausal", linestyles = "--", color = "black")

plt.xticks(size = 12)
plt.yticks(size = 12)
plt.legend([], frameon=False)
#plt.legend([],[], frameon=False)
plt.xlabel("Values", fontsize = 15)
plt.ylabel("Density of probability", fontsize = 15)
plt.title("{}".format(colonna), fontsize = 20)
fig.savefig("pics_finali/violin/three_split/medi/{}.png".format(colonna), dpi = 300)
```

# Three splitted


In questo creiamo quello donne vs fertile/menopausa in modo da evidenziare le differenze

```python
donne_and_meno
```

```python
colonna = dati.columns[17]
fig, ax = plt.subplots(figsize=(8, 6))

sns.kdeplot(donne[colonna], color='pink', ax=ax)

x = ax.lines[-1].get_xdata()
y = ax.lines[-1].get_ydata()

for p in ax.patches:  # turn the histogram upside down
    p.set_height(-p.get_height())
for l in ax.lines:  # turn the kde curve upside down
    l.set_ydata(-l.get_ydata())
    
ax.fill_between(x, 0, -y, color='pink', alpha=0.3)

sns.kdeplot(data=donne_and_meno, x=donne_and_meno[colonna], hue=donne_and_meno['menopausal_x'], shade = True, palette = ["violet","pink"])

plt.vlines(x = donne_and_meno[colonna].median(), ymin = 0, ymax = -0.1, label = "Median men", linestyles = "--", color = "black")
plt.vlines(x = donne[colonna].median(), ymin = 0, ymax = 0.08, label = "Median women", linestyles = "--", color = "black")
plt.vlines(x = menopausal[colonna].median(), ymin = 0, ymax = 0.038, label = "Median menopausal", linestyles = "--", color = "black")

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

pos_ticks = np.array([t for t in ax.get_yticks() if t > 0])
ticks = np.concatenate([-pos_ticks[::-1], [0], pos_ticks])

ax.set_yticks(ticks)
ax.set_yticklabels([f'{abs(t):.2f}' for t in ticks])
ax.spines['bottom'].set_position('zero')
plt.legend([],[], frameon=False)

plt.xlabel("Value", horizontalalignment = "right", x=1.0, size = 12)
plt.yticks(size = 12)
plt.legend([],[], frameon=False)
plt.ylabel("Density of probability", fontsize = 15)
plt.title("{}".format(colonna), fontsize = 20)
plt.show()
#fig.savefig("template.png", dpi = 300)
```
