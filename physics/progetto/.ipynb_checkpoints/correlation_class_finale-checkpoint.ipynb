# Correlation

In questo dividiamo le correlazioni per classi diverse. In particolare dividiamo il dataset in queste parti:

SEX:

- uomini
- donne
- menopausa (già prodotto sotto)

AGE:

- Donne in meno pausa
- 18-34 
- 35-50

BMI:

- <21
- 21-24.99
- \>25

Ricordiamo che questo lavoro andrà fatto sia per i valori medi che per i valori di cvi

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np 
import seaborn as sns
import math
from scipy.stats import pearsonr

print("Pandas version: ", pd.__version__)
print("Matplotlib version: ", matplotlib.__version__)
print("Seaborn version: ", sns.__version__)
print("Numpy version: ", np.__version__)

pd.options.mode.chained_assignment = None
```

```python
def calculate_pvalues(df):
    df = df.dropna()._get_numeric_data()
    dfcols = pd.DataFrame(columns=df.columns)
    pvalues = dfcols.transpose().join(dfcols, how='outer')
    for r in df.columns:
        for c in df.columns:
            pvalues[r][c] = round(pearsonr(df[r], df[c])[1], 4)
    return pvalues

def plot_cor_matrix(corr, mask=None):
    f, ax = plt.subplots(figsize=(10, 10))
    sns.heatmap(corr, ax=ax,
                mask=mask,
                # cosmetics
                annot=True, vmin=-1, vmax=1, center=0,
                cmap='coolwarm', linewidths=2, linecolor='black', cbar_kws={'orientation': 'vertical'})
    plt.xticks(size = 15)
    plt.yticks(size = 15, rotation = 0)
```

```python
dati = pd.read_csv("dati_corretti/1_rilevazione_nulli_riempiti_no_outlier.csv")
cvi = pd.read_csv("dati_corretti/cvi_finali.csv")

people = pd.read_csv("dati_corretti/people.csv", sep = ";")
```

```python
for analita in cvi.columns:
    cvi[analita] = cvi[analita].fillna(cvi[analita].mean())
```

```python
dati.rename(columns = {'Glucose': "Glu, mean values", 'Calcium': "Ca, mean values", 'Albumin': "Alb, mean values",
                      'Total Protein': "TP, mean values", 'Na': "Na, mean values", 'K': "K, mean values",
                      'Cl': "Cl, mean values", 'Urea': "Urea, mean values", 'Enz Creatinine': "Crea, mean values",
                      'ALP': "ALP, mean values", 'AST': "AST, mean values", 'ALT': "ALT, mean values", 'T Bilirubin': "TBil, mean values"}, inplace = True)
```

```python
cvi.rename(columns = {'cvi Glucose': r"Glu, $CV_P$", 'cvi Calcium': r"Ca, $CV_P$", 'cvi Albumin': r"Alb, $CV_P$",
                      'cvi Total Protein': r"TP, $CV_P$", 'cvi Na': r"Na, $CV_P$", 'cvi K': r"K, $CV_P$",
                      'cvi Cl': r"Cl, $CV_P$", 'cvi Urea': r"Urea, $CV_P$", 'cvi Enz Creatinine': r"Crea, $CV_P$",
                      'cvi ALP': r"ALP, $CV_P$", 'cvi AST': r"AST, $CV_P$", 'cvi ALT': r"ALT, $CV_P$", 
                      'cvi T Bilirubin': r"TBil, $CV_P$"}, inplace = True)
```

```python
uomini = pd.merge(people.loc[people['menopausal'] == 'male'][['id', "Age", "BMI"]], dati, on = "id")
donne = pd.merge(people.loc[people['menopausal'] == 'female'][['id', "Age", "BMI"]], dati, on = "id")
menopausal = pd.merge(people.loc[people['menopausal'] == 'menopausal female'][['id', "Age", "BMI"]], dati, on = "id")
```

```python
print("Uomini: ", len(uomini))
print("Donne: ", len(donne))
print("Menopausal: ", len(menopausal))
print("Lunghezza Totale: ", len(uomini) + len(donne) + len(menopausal))
```

## GENERALI

```python
corr = dati.drop(["id"], axis = 1).corr()      
p_values = calculate_pvalues(dati.drop(["id"], axis = 1)) 
mask = np.invert(np.tril(p_values<0.05))
plot_cor_matrix(corr,mask)
plt.savefig("pics_finali/correlation/medi/tutti_soggetti_medi.png", dpi = 300, bbox_inches='tight')
```

## CVI GENERALI

```python
corr = cvi.drop(["id"], axis = 1).corr()      
p_values = calculate_pvalues(cvi.drop(["id"], axis = 1)) 
mask = np.invert(np.tril(p_values<0.05))
plot_cor_matrix(corr,mask)
#plt.title(r"Tutti i soggetti cvi", size = 15)
plt.savefig("pics_finali/correlation/cvi/tutti_soggetti_cvi.png", dpi = 300, bbox_inches='tight')
```
